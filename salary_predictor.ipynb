```python
def extract_average(s):
    if '+' in s:
        return int(s.replace('+', ''))
    elif ' to ' in s:
        a, b = s.split(' to ')
        return (int(a) + int(b)) // 2
    else:
        return int(s)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```
